1) Diseñar una clase que permita trabajar con un árbol binario y que contenga los métodos:<br>
    * insertaVal: para insertar un dato<br>
    * buscaVal: que devuelva True o False si existe o no un dato<br>
    * verVal: que imprima por pantalla los valores del árbol<br>

Nota: Se puede utilizar la recursividad


```python
class Nodo():
    def __init__(self, dato):
        self.dato = dato
        self.izq = None
        self.der = None
```

```python
class Arbol():
    # Funciones privadas #
    # Iguala el root a la class Nodo con el dato como parametro #
    def __init__(self, dato):
        self.raiz = Nodo(dato)
    
    # Toma la entrada y el nodo (ubicacion) #
    def __insertarConRecursividad(self, nodo, dato):
        # Si el dato de entrada es menor al nodo #
        if nodo.dato > dato:
            # Ademas, si la izquierda esta vacia #
            if nodo.izq is None:
                # Nodo de la izquierda igual al dato #
                nodo.izq = Nodo(dato)
            else:
            # Recursividad con el nodo de la izquierda y el dato com parametros#
            # Mientras el nodo de la izquierda sea mayor al dato se buscara un espacio más hacia la izquierda #
                self.__insertarConRecursividad(nodo.izq, dato)
        # Si es menor #
        else:
            # El mismo proceso, hacia la derecha #
            if nodo.der is None:
                nodo.der = Nodo(dato)
            else:
                self.__insertarConRecursividad(nodo.der, dato)
    
    def __inordenRecursividad(self, nodo):
        # Si el nodo existe #
        if nodo is not None:
            # Imprime el dato de la izquierda #
            self.__inordenRecursividad(nodo.izq)
            # Imprime el dato del nodo #
            print(nodo.dato)
            # Imprime el dato de la derecha #
            self.__inordenRecursividad(nodo.der)

    def __preordenRecursividad(self, nodo):
        # Si el nodo existe #
        if nodo is not None:
            # Imprime el dato del nodo #
            print(nodo.dato)
            # Imprime el dato de la izquierda #
            self.__preordenRecursividad(nodo.izq)
            # Imprime el dato de la derecha #
            self.__preordenRecursividad(nodo.der)

    def __postordenRecursividad(self, nodo):
        # Si el nodo existe #
        if nodo is not None:
            # Imprime el dato de la izquierda #
            self.__postordenRecursividad(nodo.izq)
            # Imprime el dato de la derecha #
            self.__postordenRecursividad(nodo.der)
            # Imprime el dato del nodo #
            print(nodo.dato)
    
    def __buscar(self, nodo, busqueda):
        # Si el nodo existe #
        if nodo is not None:
            # Si el dato del nodo es igual al dato de busqueda #
            if nodo.dato == busqueda:
                # Retorna el nodo #
                return nodo
            # Si el dato del nodo es mayor al dato de busqueda #
            elif nodo.dato > busqueda:
                # Retorna el nodo de la izquierda #
                return self.__buscar(nodo.izq, busqueda)
            # Si el dato del nodo es menor al dato de busqueda #
            else:
                # Retorna el nodo de la derecha #
                return self.__buscar(nodo.der, busqueda)
        # Si el nodo no existe #
        else:
            # Retorna None #
            return None

    # Funciones publicas #
    def Agregar(self, dato):
        # Inserta el dato en el arbol #
        self.__insertarConRecursividad(self.raiz, dato)

    def Inorden(self):
        # Imprime el arbol en inorden #
        print("Inorden: ")
        self.__inordenRecursividad(self.raiz)
        print('')

    def Preorden(self):
        # Imprime el arbol en preorden #
        print("Preorden: ")
        self.__preordenRecursividad(self.raiz)
        print('')

    def Postorden(self):    
        # Imprime el arbol en postorden #
        print("Postorden: ")
        self.__postordenRecursividad(self.raiz)
        print('')

    def Buscar(self, busqueda):
        # Busca el dato en el arbol #
        return self.__buscar(self.raiz, busqueda)
    

```

```python
arbol = Arbol('Luis')
arbol.Agregar('Juan')
arbol.Agregar('Pedro')
arbol.Agregar('Carlos')
arbol.Agregar('Jorge')
arbol.Agregar('Juan')
```

```python
arbol.Inorden()
arbol.Preorden()
arbol.Postorden()
```
